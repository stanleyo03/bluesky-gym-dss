```python
import pandas as pd
import matplotlib
import seaborn as sns

%matplotlib inline

train_df1 = pd.read_csv("logs\DescentEnvXYZ-v0\DescentEnvXYZ-v0_SAC-2.csv")
train_df2 = pd.read_csv("logs\DescentEnvXYZ-v0\DescentEnvXYZ-v0_SAC-50k-100k.csv")
train_df2["timesteps"] = train_df2["timesteps"] + 50000
```

```python
train_df = pd.concat([train_df1, train_df2])

train_df.head()
```

```python
sns.lineplot(train_df, x="timesteps", y = "total_reward")
```

```python
sns.lineplot(train_df[train_df["timesteps"] > 500], x="timesteps", y = "total_reward")
```
